```python
import numpy as np
import matplotlib.pyplot as plt
import math as m # Import math library for pi
```

```python
# Parameters
k = 5  # Force constant for harmonic potential
r0 = 1.0  # Equilibrium distance
sigma = 1.0  # Standard deviation for Gaussian
epsilon = 5  # Depth of potential well for Gaussian

# Distance range for plotting
r = np.linspace(0.0, 3.0, 500)  # From 0.5 to 2.5 for visualization

# Harmonic potential
V_harmonic = 0.5 * k * (r - r0)**2

# Gaussian potential
V_gaussian = epsilon * np.exp(-((abs(r)**2) / (2 * sigma**2)))

# Sum of potentials
V_sum = V_harmonic + V_gaussian
```

```python
# Print the minimum of the sum

# Plotting
plt.figure(figsize=(10, 10))
plt.plot(r, V_harmonic, label='Harmonic')
plt.plot(r, V_gaussian, label='Gaussian')
plt.plot(r, V_sum, label='Sum', linestyle='--')
plt.xlim(0.0, 3.0)
plt.xlabel(r'$r_{ij}$')
plt.ylabel(r'$U(r_{ij})$')
#Find r for minimum of the sum
r_eq = r[np.argmin(V_sum)]
# Add vertical line at the minimum of the sum 
plt.axvline(x=r_eq, color='k', linestyle='--')
# label it
plt.text(r_eq, 0.5, r'$r_{eq}$', fontsize=12)


# Increase font size globally
plt.rcParams['font.size'] = 12  # Adjust this value as needed
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['legend.fontsize'] = 18
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Parameters
kappa = 1  # Bending rigidity constant, arbitrary units

# Angle range from 0 to pi radians
theta = np.linspace(0, np.pi, 100)

# Wormlike chain angular potential
V_theta = kappa * (1 + np.cos(theta))


# Plotting
plt.figure(figsize=(10,10))
plt.plot(theta, V_theta, label='WLC Angular Potential')
plt.xlabel(r'$\theta_{ijk}$ (radians)')
plt.ylabel(r'$u(\theta_{ijk})$')
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['legend.fontsize'] = 18
plt.legend()
plt.grid(True)
plt.show()
```
